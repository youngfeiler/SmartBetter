```python
import numpy as np
import random
import matplotlib.pyplot as plt
import statistics
```

```python
def roll(bet):

    odds_to_win = (.9 * 100)

    rand_num = random.randint(1,100)

    if rand_num <= odds_to_win:
        return (bet*1.1)
    else:
        return bet*-1
```

```python
def change_bet(streak, bet):
    initial_bet = 1
    if streak[-3:] == [1, 1, 1]:
      return initial_bet
    else:
       return bet*2
```

```python
def iterate():
  
  initial_bet = 1
  
  bet = initial_bet

  num_rolls = 50_000
  
  bankroll = 1000
  
  streak = []
  
  rolls = []
  
  bankrolls = []
  
  for num in range(num_rolls):
      
      rolls.append(num)

      p_l = roll(bet)

      bankroll += p_l

      if bankroll < 0 :
          return 0, rolls, bankrolls

      if p_l > 0:
          streak.append(1)

      elif p_l < 0:
          streak.append(0)
      
      bet = change_bet(streak, bet)

      bankrolls.append(bankroll)

  return bankroll, rolls, bankrolls
    
```

```python
final, rolls, bankrolls = iterate()
print(final)
```

```python
# plot scatter plot
plt.scatter(rolls, bankrolls)

# set axis labels and title
plt.xlabel('Roll #')
plt.ylabel('Running Bankroll')
plt.title('Bankroll over time')

# show plot
plt.show()
```

```python
ending_brs = []
for time_ran in range(1000):
    bankroll, rolls, bankrolls = iterate()
    ending_brs.append(bankroll)
```
